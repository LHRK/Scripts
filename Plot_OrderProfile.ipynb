```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_ref_profiles(system):
    P = []
    for r in range(1,4):
        P.append(np.loadtxt('ORDER/{0:s}_v{1:d}_nowat-order-profile.xvg'.format(system,r), comments=('#','@')))
    return P
```

```python
pro_1D1 = load_ref_profiles('1D1_69')
```

```python
pro_1E3D1 = load_ref_profiles('1E3D1_134')
```

```python
pro_2N2 = load_ref_profiles('2N2_302')
```

```python
fig, ax = plt.subplots(nrows=1, ncols=3, sharey=True, figsize=(15,5))
[ ax[0].plot(pro_1D1[r][:,0], pro_1D1[r][:,1], label='Replica {0:d}'.format(r+1)) for r in range(3) ] 
[ ax[1].plot(pro_1E3D1[r][:,0], pro_1E3D1[r][:,1], label='Replica {0:d}'.format(r+1)) for r in range(3) ] 
[ ax[2].plot(pro_2N2[r][:,0], pro_2N2[r][:,1], label='Replica {0:d}'.format(r+1)) for r in range(3) ] 

systems = ['1D1', '1E3D1', '2N2']

[ ax[r].set_title('{0:s}'.format(systems[r]), fontsize=14) for r in range(3) ]

[ ax[i].grid() for i in range(3) ]
plt.legend(loc='center left', bbox_to_anchor=(1, 0.5), frameon=False, fontsize=13)

[ ax[i].set_xlabel('x-plane (nm)', fontsize=13) for i in range(3) ]
ax[0].set_ylabel('Second Order parameter', fontsize=13) 
ax[0].set_ylim(-0.55, 0.8)

plt.tight_layout()

plt.savefig('Order_profiles_Sligar.png', dpi=600)
```
