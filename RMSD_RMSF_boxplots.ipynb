```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def load_data_rmsd(name):
    d1 = np.loadtxt('RMSD/{0:s}_v1_nowat_BB_rmsd.xvg'.format(name), comments=('#','@'))[:,1]
    d2 = np.loadtxt('RMSD/{0:s}_v2_nowat_BB_rmsd.xvg'.format(name), comments=('#','@'))[:,1]
    d3 = np.loadtxt('RMSD/{0:s}_v3_nowat_BB_rmsd.xvg'.format(name), comments=('#','@'))[:,1]
    D  = np.concatenate((d1, d2, d3))
    return D
```

```python
def load_data_rmsf(name):
    d1 = np.loadtxt('RMSF/{0:s}_v1_nowat_BB_rmsf.xvg'.format(name), comments=('#','@'))[:,1]
    d2 = np.loadtxt('RMSF/{0:s}_v2_nowat_BB_rmsf.xvg'.format(name), comments=('#','@'))[:,1]
    d3 = np.loadtxt('RMSF/{0:s}_v3_nowat_BB_rmsf.xvg'.format(name), comments=('#','@'))[:,1]
    D  = np.concatenate((d1, d2, d3))
    return D
```

```python
systems = ['1D1_60', '1E3D1_134', 'NW9_54', 'NW11_70', 'NWLA_150']
```

```python
data_rmsd = []
data_rmsf = []
for s in systems:
    data_rmsd.append(load_data_rmsd(s))
    data_rmsf.append(load_data_rmsf(s))
```

```python
data_rmsd_array = np.array(data_rmsd)
data_rmsf_array = np.array(data_rmsf)
```

```python
plt.figure(figsize=(12,4))
plt.boxplot(data_rmsd_array, 0, '', labels=('1D1 #60 POPC\n Non-circularized \n(~9.7 nm)','1E3D1 #134 POPC\n Non-circularized\n(~13 nm)', 'NW9 #54 POPC\n Circularized \n(~9 nm)', 'NW11 #70 POPC\n Circularized \n(~11 nm)', 'NW13 #150 POPC\n Circularized \n(~13 nm)'), showbox=False)
plt.ylabel('RMSD (nm)', fontsize=13)
plt.xticks(fontsize=13)
plt.yticks(fontsize=13)
plt.tight_layout()
plt.ylim(0, 2.5)
plt.savefig('RMSD_boxplot.png', dpi=600)
```

```python
plt.figure(figsize=(12,4))
plt.boxplot(data_rmsf_array, 0, '', labels=('1D1 #60 POPC\n Non-circularized \n(~9.7 nm)','1E3D1 #134 POPC\n Non-circularized\n(~13 nm)', 'NW9 #54 POPC\n Circularized \n(~9 nm)', 'NW11 #70 POPC\n Circularized \n(~11 nm)', 'NW13 #150 POPC\n Circularized \n(~13 nm)'), showbox=False)
#plt.boxplot(data_rmsf_array, 0, '', labels=('1D1 #60 POPC\n Non-circularized \n(~9.7 nm)','1D1 #65 POPC\n Non-circularized \n(~9.7 nm)','1D1 #80 POPC\n Non-circularized \n(~9.7 nm)','1E3D1 #134 POPC\n Non-circularized\n(~13 nm)', 'NW9 #54 POPC\n Circularized \n(~9 nm)', 'NW11 #70 POPC\n Circularized \n(~11 nm)', 'NW13 #150 POPC\n Circularized \n(~13 nm)'), showbox=False)
plt.ylabel('RMSF (nm)')
plt.tight_layout()
plt.ylim(0,1.6)
plt.savefig('RMSF_boxplot.png', dpi=800)
```
