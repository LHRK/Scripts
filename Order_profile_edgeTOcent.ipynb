```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
systems_1D1 = ['1D1_60', '1D1_65', '1D1_70', '1D1_75', '1D1_80', '1D1_85', '1D1_90']
systems_NW11 = ['NW11_70', 'NW11_80', 'NW11_90', 'NW11_100', 'NW11_110', 'NW11_119']
```

```python
D_1D1 = []
for s in systems_1D1:
    D_1D1.append(np.loadtxt('ORDER/{0:s}_v1_nowat-order-profile.xvg'.format(s), skiprows=25))
    
D_NW11 = []
for s in systems_NW11:
    D_NW11.append(np.loadtxt('ORDER/{0:s}_v1_nowat-order-profile.xvg'.format(s), skiprows=25))
```

```python
def plot(d, label_txt, color, xmin, xmax):
    ax.plot(d[:,0], d[:,1], label=label_txt, color = color)
    ax.grid(True)
    ax.set_xlim(xmin,xmax)
    ax.set_xlabel('Z')
    ax.set_ylabel('Order Parameter')
    return
```

```python
def plot2(d, label_txt, color, xmin, xmax):
    f, ax = plt.subplots()
    ax.plot(d[:,0], d[:,1], label=label_txt, color = color)
    ax.grid(True)
    ax.set_xlim(xmin,xmax)
    ax.set_xlabel('Z')
    ax.set_ylabel('Order Parameter')
    return
```

```python
color_list = ['#a6cee3', '#1f78b4', '#b2df8a', '#33a02c', '#fb9a99', '#e31a1c', '#fdbf6f']
f, ax = plt.subplots()
for idx, d in enumerate(D_1D1):
    plot(d, systems_1D1[idx], color_list[idx], 5, 15)
    plt.legend(loc='center left', bbox_to_anchor=(1, 0.5))
```
